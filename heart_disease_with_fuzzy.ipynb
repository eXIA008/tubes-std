# Fuzzy Logic

## a. Fungsi Keanggotaan

```python
# Fungsi Keanggotaan untuk age

def age_muda(usia):
    if usia <= 30:
        return 1
    elif usia <= 45:
        return (45 - usia) / (45 - 30)
    else:
        return 0

def age_sedang(usia):
    if usia <= 40 or usia >= 65:
        return 0
    elif 40 < usia <= 52.5:
        return (usia - 40) / (52.5 - 40)
    elif 52.5 < usia < 65:
        return (65 - usia) / (65 - 52.5)

def age_tua(usia):
    if usia <= 60:
        return 0
    elif 60 < usia <= 70:
        return (usia - 60) / (70 - 60)
    else:
        return 1

# Fungsi Keanggotaan untuk chol
def cholesterol_normal(kol):
    if kol <= 160:
        return 1
    elif kol <= 200:
        return (200 - kol) / (200 - 160)
    else:
        return 0

def cholesterol_tinggi(kol):
    if kol <= 180 or kol >= 300:
        return 0
    elif 180 < kol <= 240:
        return (kol - 180) / (240 - 180)
    elif 240 < kol < 300:
        return (300 - kol) / (300 - 240)

def cholesterol_sangat_tinggi(kol):
    if kol <= 280:
        return 0
    elif 280 < kol <= 400:
        return (kol - 280) / (400 - 280)
    else:
        return 1
```

## d. Inference

```python
def fuzzy_inference(age, chol):
    muda = age_muda(age)
    sedang_age = age_sedang(age)
    tua = age_tua(age)

    normal = cholesterol_normal(chol)
    tinggi = cholesterol_tinggi(chol)
    sangat_tinggi = cholesterol_sangat_tinggi(chol)

    # Aturan Fuzzy
    rules = {
        "rendah" : [
            min(muda, normal),
            min(sedang_age, normal),
        ],
        "sedang" : [
            min(muda, tinggi),
            min(muda, sangat_tinggi),
            min(sedang_age, tinggi),
            min(tua, normal)
        ],
        "tinggi" : [
            min(sedang_age, sangat_tinggi),
            min(tua, tinggi),
            min(tua, sangat_tinggi)
        ]
    }
    rendah = max(rules["rendah"])
    sedang = max(rules["sedang"])
    tinggi = max(rules["tinggi"])

    return {"Tinggi": tinggi, "Sedang": sedang, "Rendah": rendah}
```

## e. Defuzzyfikasi


```python
# Fungsi Keanggotaan Output Fuzzy
def risiko_rendah(x):
    if x <= 20:
        return 1
    elif 20 < x <= 40:
        return (40 - x) / (40 - 20)
    else:
        return 0

def risiko_sedang(x):
    if x <= 30 or x >= 70:
        return 0
    elif 30 < x <= 50:
        return (x - 30) / (50 - 30)
    elif 50 < x < 70:
        return (70 - x) / (70 - 50)
    else:
        return 0

def risiko_tinggi(x):
    if x <= 60:
        return 0
    elif 60 < x <= 80:
        return (x - 60) / (80 - 60)
    elif 80 < x <= 100:
        return 1
    else:
        return 0
```
